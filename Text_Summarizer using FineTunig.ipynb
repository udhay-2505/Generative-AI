```python
tokenizer= AutoTokenizer.from_pretrained("bert-base-cased")
text="i was not happy with the last movie Mission Impossible that i watched"
tokens=tokenizer.tokenize(text)
print(tokens)
```

```python
input_ids=tokenizer.convert_tokens_to_ids(tokens)
print(input_ids)
```

```python
encoded_ids = tokenizer(text)
print("endcoded ids:",encoded_ids)
```

```python
encoded_ids = tokenizer.encode(text)
print("endcoded ids:",encoded_ids)
decoded_text=tokenizer.decode(input_ids)
print("decoded text:",decoded_text)
```

FINE TUNING USING PRE TRAINED MODELS


```python
!pip install datasets
```

```python
from datasets import load_dataset
dataset = load_dataset('stanfordnlp/imdb')
```

```python
dataset
```

```python
dataset["train"][2]
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

def tokenize_function(examples):
    return tokenizer(examples['text'], padding="max_length", truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
tokenized_datasets["train"][2]
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir='./results',
    eval_strategy ="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=1,
    weight_decay=0.01,
)
training_args
```

```python
from transformers import AutoModelForSequenceClassification, Trainer

# Load the pre-trained model
model = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['test']
)
```

```python
trainer.train()
```

```python
results = trainer.evaluate()
print(results)
```

```python
model.save_pretrained('./fine-tuned-model')
tokenizer.save_pretrained('./fine-tuned-tokenizer')
```

```python
import arxiv
import pandas as pd
```

```python
query = 'ai OR artificial intelligence OR machine learning'
search = arxiv.Search(query=query, max_results=10, sort_by=arxiv.SortCriterion.SubmittedDate)

# Fetch papers
papers = []
for result in search.results():
    papers.append({
      'published': result.published,
        'title': result.title,
        'abstract': result.summary,
        'categories': result.categories
    })

# Convert to DataFrame
df = pd.DataFrame(papers)

pd.set_option('display.max_colwidth', None)
df.head(10)
```

```python
abstract= df['abstract'][0]

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
SummarizationOutput=summarizer(abstract)
```
